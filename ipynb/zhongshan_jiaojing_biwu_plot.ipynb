```python
data = pd.read_csv("/data/git/study-note/data/result2.txt",sep = "," , header = None, 
                   names = ["date_period","date_type","step_index","date_time",
                            "cong_index","u_index","delay_index","com_index"])
weeks = data["date_period"].unique()
data=data.sort_values(by = ["date_period","date_type","step_index"] ,ascending=[True,True,True])
type(weeks)
type(data["cong_index"])
test = data.loc[(data["date_period"] == "0325-0331") & (data["date_type"] == "holi_date") ]
```

```python
ali_data = data.loc[(data["date_period"] == "0325-0331") & (data["date_type"] == "holi_date") ]
qingsong_data = data.loc[(data["date_period"] == "0408-0414") & (data["date_type"] == "holi_date") ]
huagong_data = data.loc[(data["date_period"] == "0422-0428") & (data["date_type"] == "holi_date") ]
test_data = data.loc[(data["date_period"] == "0508-0514") & (data["date_type"] == "holi_date") ]

test_data
```

```python
step_index = np.arange(144)
date_time = test_data["date_time"]


plt.plot(step_index, test_data["u_index"], color = 'green', linestyle = '--', marker = '+', linewidth = 1,label = 'test data')
plt.plot(step_index, ali_data["u_index"], color = 'red', linestyle = '-', marker = '*', linewidth = 2,label = 'ali data')
plt.plot(step_index, qingsong_data["u_index"], color = 'blue', linestyle = '-', marker = '+', linewidth = 1,label = 'qingsong data')
plt.plot(step_index, huagong_data["u_index"], color = 'black', linestyle = '-', marker = '+', linewidth = 1,label = 'huagong data')
plt.xticks(step_index, date_time,rotation = 90)
plt.legend(['test data', 'ali data', 'qingsong data', 'huagong data'], loc='upper left')
plt.title('test')
plt.autoscale(tight=True)
fig =plt.gcf()
fig.set_size_inches(25, 7.5)
plt.xlabel(u"")
plt.ylabel(u"")
plt.show()
```
